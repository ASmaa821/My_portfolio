```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned_data.csv")
df
```

```python


cols = ["total_sales", "north_america_sales", "europe_sales","other_sales"]
df_melted = df[cols].melt(var_name="Region", value_name="Sales")
sns.boxplot(data=df_melted, x="Region", y="Sales")
plt.title("Boxplot of Sales by Region")
plt.show()
```

```python
# ŸÜÿ≠ÿØÿØ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑŸÑŸä ŸÅŸäŸáÿß ŸÖÿ®Ÿäÿπÿßÿ™
sales_columns = ['north_america_sales', 'japan_sales', 'europe_sales', 'other_sales']

for col in sales_columns:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1

  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  df=df[(df[col] > lower_bound) & (df[col] < upper_bound)]



  print(len(outliers))
  print(f"lower bound {lower_bound}:")
  print(f"upper bound {upper_bound}:")
```

```python

sales_cols = ['north_america_sales', 'japan_sales', 'europe_sales','total_sales']
all_outlier_indices = set()

for col in sales_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(outliers.head())
    outlier_indices = set(outliers.index)
    all_outlier_indices = all_outlier_indices.union(outlier_indices)

```

```python

sales_cols = ['north_america_sales', 'japan_sales', 'europe_sales','total_sales']

grouped_sales = df.groupby('title')[sales_cols].sum().sort_values(by='total_sales', ascending=False)

print(grouped_sales.head(10))
```

```python
top_publishers = df.groupby('publisher')['total_sales'].sum().sort_values(ascending=False).head(10)
top_publishers.plot(kind='bar', title='Top 10 Publishers by Total Sales')
```

```python
pivot = df.groupby(['consol', 'Genre'])['total_sales'].mean().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu")

plt.show()
```

```python
top10 = df.sort_values(by='total_sales', ascending=False).head(10)
top10.set_index('title')[['north_america_sales', 'japan_sales', 'europe_sales', 'other_sales']].plot(kind='bar', figsize=(15, 6))
plt.title("Top 10 Games Sales by Region")
```

```python
regions = ['north_america_sales', 'japan_sales', 'europe_sales', 'other_sales']
region_sales = df[regions].sum()

region_sales.plot.pie(autopct='%1.1f%%', figsize=(7,7), startangle=140)

plt.show()
```
